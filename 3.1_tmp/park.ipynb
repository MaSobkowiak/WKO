```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
park_raw = cv.imread("park-to-eq.jpg", cv.IMREAD_COLOR)

plt.figure(figsize=(10,10))
plt.imshow(park_raw[:,:,::-1])
plt.title("Image");
```

```python
R, G, B = cv.split(park_raw)
R = cv.equalizeHist(R) 
G = cv.equalizeHist(G)
B = cv.equalizeHist(B)
park_restored = cv.merge((R,G,B))
plt.figure(figsize=(10,10))
plt.imshow(park_restored[:,:,::-1])
plt.title("Image");
```

```python
park_hsv = cv.cvtColor(park_raw, cv.COLOR_BGR2HSV)
H, S, V = cv.split(park_hsv)

# H = cv.equalizeHist(H) 
# S = cv.equalizeHist(S)
V = cv.equalizeHist(V)

park_hsv_restored = cv.merge((H,S,V))

park_rgb_restored = cv.cvtColor(park_hsv_restored, cv.COLOR_HSV2BGR)

plt.figure(figsize=(10,10))
plt.imshow(park_rgb_restored[:,:,::-1])
plt.title("Image");
```
