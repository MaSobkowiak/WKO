```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
man = cv.imread("man-at-desk.jpg", cv.IMREAD_COLOR)

plt.figure(figsize=(10,10))
plt.imshow(man[:,:,::-1])
plt.title("Image");
```

```python
man_hsv = cv.cvtColor(man, cv.COLOR_BGR2HSV)
H, S, V = cv.split(man_hsv)

# H = cv.equalizeHist(H) 
# S = cv.equalizeHist(S)
V = cv.equalizeHist(V)

man_hsv_restored = cv.merge((H,S,V))

man_rgb_restored = cv.cvtColor(man_hsv_restored, cv.COLOR_HSV2BGR)

plt.figure(figsize=(10,10))
plt.imshow(man_rgb_restored[:,:,::-1])
plt.title("Image");
```

```python
man_hsv_clahe = cv.cvtColor(man, cv.COLOR_BGR2HSV)
H, S, V = cv.split(man_hsv_clahe)

clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(5,5))
V = clahe.apply(V)

man_hsv_restored = cv.merge((H,S,V))

man_rgb_restored_clahe = cv.cvtColor(man_hsv_restored, cv.COLOR_HSV2BGR)

plt.figure(figsize=(10,10))
plt.imshow(man_rgb_restored_clahe[:,:,::-1])
plt.title("Image");
```

```python
plt.figure(figsize=(20,40))
plt.subplot(131)
plt.imshow(man[:,:,::-1])
plt.title("Orginal")
plt.subplot(132)
plt.imshow(man_rgb_restored[:,:,::-1])
plt.title("eqalizerHist")
plt.subplot(133)
plt.imshow(man_rgb_restored_clahe[:,:,::-1])
plt.title("CLAHE")
```
